### Dataset description

In this project, I mainly use three datasets from the Centers for Disease Control and Prevention (CDC) official website. One is the ["COVID-19 Vaccinations in the United States, Jurisdiction"](https://data.cdc.gov/Vaccinations/COVID-19-Vaccinations-in-the-United-States-Jurisdi/unsk-b7fc), which has more than 23 thousand rows and 80 columns with variables I choose to use including date ("Date"), state ("Location"), percentage of fully vaccinated people ("Series_Complete_Pop_Pct"), and percentage of people who received at least one shot ("Administered_Dose1_Pop_Pct"). It records the vaccination data of the US in state level since 12/13/2020. The second is ["COVID-19 Vaccinations in the United States, County"](https://data.cdc.gov/Vaccinations/COVID-19-Vaccinations-in-the-United-States-County/8xkx-amqh), which has more than 1.19 million rows and 32 columns with variables I choose to use including date ("Date"), state ("Recip_State"), county ("Recip_County"), percentage of fully vaccinated people ("Series_Complete_Pop_Pct"), and percentage of people who received at least one shot ("Administered_Dose1_Pop_Pct"). It records the vaccination data of the US in county level since 12/13/2020. The last is ["United States COVID-19 County Level of Community Transmission as Originally Posted"](https://data.cdc.gov/Public-Health-Surveillance/United-States-COVID-19-County-Level-of-Community-T/8396-v7yb), which has more than 374 thousand rows and 7 columns with variables I choose to use including date ("report_date"), state ("state_name"), county ("county_name"), number of new cases per 100 thousand people within 7 days ("cases_per_100K_7_day_count_change"), percentage of positive tests within 7 days ("percent_test_results_reported_positive_last_7_days"), and so on. The data is also recorded in county level since 08/16/2021. All three datasets are keep updating data until the latest date as well as my web-app.

### Question of interest  

In this project, I'm targeting to develope an interactive web-app with "Dash" combined with data manipulation and aggregation technics within "pandas" package, and use functions from "plotly" package to display visualizations of the results. This web-app is used to track the covid-19 data in the United States to state and county level according to the information given by CDC. It is expected to show one graph of the map of the entire United States with each states colored by either percentage of fully vaccinated people or percentage of people who received at least one shot for the date selected by user. It also has a table displaying the vaccination rate data from all counties of user selected state sorting by the user's preference in selected date. Furthermore, the app contains two line charts of 7-day moving average of positive test percentage and the number of daily new cases in selected county within user defined time range.
